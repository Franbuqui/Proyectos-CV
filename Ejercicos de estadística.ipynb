# 1 
Una de las enfermedades que más preocupa a las autoridades sanitarias es 
la hipertensión. La hipertensión arterial afecta más del 30% de la población 
adulta mundial y es el principal factor de riesgo de padecer enfermedades 
cardiovasculares. Por ello, la lucha contra esta enfermedad es objetivo de los 
sistemas sanitarios de todo el mundo.                                                                                    
En un laboratorio europeo se está trabajando en un medicamento innovador 
con cierta composición de etanol para reducir la presión arterial alta 
(sistólica). El medicamento se suministra a 100 pacientes de diferentes 
edades por vía oral en forma de comprimido. Para obtener mejores 
conclusiones en el estudio, se divide a las 100 personas en dos grupos, grupo 
1 (pacientes con edad igual o inferior a 35 años) y grupo 2 (pacientes con 
edad superior a 35 años). En la hoja de cálculo Excel llamada “HIPERTENSIÓN 
“aparecen los datos de las 100 personas que se han empleado en el estudio. 
En la primera columna aparece la edad del paciente, en la segunda columna 
aparece la presión sistólica antes de la ingestión del medicamento y en la 
tercera columna aparece indicado el valor de la presión sistólica 60 minutos 
después de la toma del medicamento. Los 40 primeros datos corresponden a 
pacientes del primer grupo y los 60 restantes a pacientes del segundo grupo. 
También se incluye una última columna donde figura el valor del colesterol 
total del paciente. 
A partir de los datos que aparecen en la hoja de cálculo, se pide contestar a 
las siguientes cuestiones:

a) Obtener, usando algún programa estadístico, las medidas de 
centralización y dispersión para cada uno de los dos grupos de control de 
la variable (grupo 1 y grupo 2) que mide la presión sistólica antes de la 
toma del medicamento e indica si la media en cada uno de estos grupos 
puede considerarse representativa a partir de los datos obtenidos. 

b) Estudiar la simetría y la curtosis del nivel de presión sistólica en los 
pacientes del segundo grupo para cada una de las mediciones de la 
hipertensión que aparecen en la tabla. 

c) Indicar para cada una de las variables relacionadas con la medición 
de la presión sistólica que aparecen en el fichero el valor de los cuartiles 
y su significado y obtener el box- plot ( diagrama de cajas) 
correspondiente. Estudiar la presencia de valores atípicos 

d) Estudiar la normalidad de los datos de las variables relacionadas con 
la medición de la presión sistólica. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, mode
from statsmodels.stats.stattools import jarque_bera
import seaborn as sns

# Cargar archivo Excel con los datos de Hipertension
df = pd.read_excel("HIPERTENSION.xlsx")

# Dividir en dos grupos según la edad
grupo1 = df[df["Edad"] <= 35]
grupo2 = df[df["Edad"] > 35]

#Ejercicio 1

#a)
print("a)")
# Medidas de centralización y dispersión para grupo 1
media_grupo1 = grupo1["Presion sistólica antes"].mean()
mediana_grupo1 = grupo1["Presion sistólica antes"].median()
desviacion_estandar_grupo1 = grupo1["Presion sistólica antes"].std()
varianza_grupo1 = grupo1["Presion sistólica antes"].var()
moda_grupo1 = mode(grupo1["Presion sistólica antes"])

# Medidas de centralización y dispersión para grupo 2
media_grupo2 = grupo2["Presion sistólica antes"].mean()
mediana_grupo2 = grupo2["Presion sistólica antes"].median()
desviacion_estandar_grupo2 = grupo2["Presion sistólica antes"].std()
varianza_grupo2 = grupo2["Presion sistólica antes"].var()
moda_grupo2 = mode(grupo2["Presion sistólica antes"])

def calcular_modas(data):
    counts = np.bincount(data)
    max_count = np.max(counts)
    modas = np.where(counts == max_count)[0]
    return modas

modas_grupo1 = calcular_modas(grupo1["Presion sistólica antes"])
modas_grupo2 = calcular_modas(grupo2["Presion sistólica antes"])

print("Medidas de centralización y dispersión para Grupo 1:")
print("Media:", media_grupo1)
print("Mediana:", mediana_grupo1)
print("Desviación estándar:", desviacion_estandar_grupo1)
print("Varianza:", varianza_grupo1)
print("Modas:", modas_grupo1)
print("\n")

print("Medidas de centralización y dispersión para Grupo 2:")
print("Media:", media_grupo2)
print("Mediana:", mediana_grupo2)
print("Desviación estándar:", desviacion_estandar_grupo2)
print("Varianza:", varianza_grupo2)
print("Modas:", modas_grupo2)
print("\n")

print("La diferencia entre la media y la mediana es pequeña (1,075), lo que sugiere que la media es una representación razonable de la distribución de los datos en Grupo 1.")
print("La diferencia entre la media y la mediana es también relativamente pequeña (1,483), lo que sugiere que la media es una representación razonable de la distribución de los datos en Grupo 2.")
print("\n")

#b)

print("b)")
# Calcular simetría y curtosis para Presion sistólica antes en Grupo 2
print("Simetría y curtosis del grupo 2 para Presion sistólica antes:")
print("Simetría:", grupo2['Presion sistólica antes'].skew())
print("Curtosis:", grupo2['Presion sistólica antes'].kurtosis())
print("\n")

# Calcular simetría y curtosis para Presion sistólica después en Grupo 2
print("Simetría y curtosis del grupo 2 para Presion sistólica después:")
print("Simetría:", grupo2['Presión sistólica después'].skew())
print("Curtosis:", grupo2['Presión sistólica después'].kurtosis())
print("\n")

print("En cuanto la presión sistólica antes en el Grupo 2, la simetría es aproximadamente -0.137, lo que indica una ligera asimetría hacia la izquierda, y la curtosis es aproximadamente -0.602, lo que indica una distribución relativamente aplanada en comparación con una distribución normal(platicúrtica).")
print("La presión sistólica después en el Grupo 2, la simetría es aproximadamente -0.194, lo que también indica una ligera asimetría hacia la izquierda, y la curtosis es aproximadamente 0.369, lo que indica que la distribución es más puntiaguda y concentrada alrededor de la media en comparación con una distribución normal(leptocúrtica).")
print("\n")

#c)
print("c)")
# Variables relacionadas con la medición de la presión sistólica
variables_presion_sistolica = ["Presion sistólica antes", "Presión sistólica después"]

# Obtener cuartiles y realizar boxplot para cada variable
for variable in variables_presion_sistolica:
    # Cuartiles
    cuartiles = df[variable].quantile([0.25, 0.5, 0.75])
    print(f"Cuartiles para {variable}:")
    print(cuartiles)
    
    # Boxplot
    plt.figure(figsize=(8, 6))
    plt.title(f"Boxplot para {variable}")
    plt.boxplot(df[variable])
    plt.ylabel("Presión Sistólica")
    plt.show()
    
    # Identificar valores atípicos
    Q1 = cuartiles[0.25]
    Q3 = cuartiles[0.75]
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    valores_atipicos = df[(df[variable] < limite_inferior) | (df[variable] > limite_superior)][variable]
    if len(valores_atipicos) > 0:
        print(f"Valores atípicos para {variable}:")
        print(valores_atipicos)
    else:
        print(f"No hay valores atípicos para {variable}")
    print("\n")

    
#d)
print("d)")
# Realizar la prueba de normalidad para cada variable
for variable in variables_presion_sistolica:
    # Realizar la prueba de Shapiro-Wilk
    stat_sw, p_value_sw = shapiro(df[variable])
    
    # Realizar la prueba de Kolmogorov-Smirnov
    stat_ks, p_value_ks = kstest(df[variable], 'norm')
    
    # Imprimir resultados de Shapiro-Wilk
    print(f"Prueba de Shapiro-Wilk para {variable}:")
    print(f"Estadístico de prueba: {stat_sw}")
    print(f"Valor p: {p_value_sw}")
    
    # Interpretar el valor p de Shapiro-Wilk
    alpha = 0.05
    if p_value_sw > alpha:
        print("No se rechaza la hipótesis nula: Los datos parecen provenir de una distribución normal según Shapiro-Wilk.")
    else:
        print("Se rechaza la hipótesis nula: Los datos no parecen provenir de una distribución normal según Shapiro-Wilk.")
    
    # Imprimir resultados de Kolmogorov-Smirnov
    print(f"\nPrueba de Kolmogorov-Smirnov para {variable}:")
    print(f"Estadístico de prueba: {stat_ks}")
    print(f"Valor p: {p_value_ks}")
    
    # Interpretar el valor p de Kolmogorov-Smirnov
    if p_value_ks > alpha:
        print("No se rechaza la hipótesis nula: Los datos parecen provenir de una distribución normal según Kolmogorov-Smirnov.")
    else:
        print("Se rechaza la hipótesis nula: Los datos no parecen provenir de una distribución normal según Kolmogorov-Smirnov.")
    
    print("\n")

# Datos y ajuste de la distribución normal
datos_antes = grupo2['Presion sistólica antes']
datos_despues = grupo2['Presión sistólica después']
mu_antes, std_antes = datos_antes.mean(), datos_antes.std()
mu_despues, std_despues = datos_despues.mean(), datos_despues.std()

# Crear figuras y ejes
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Histograma y ajuste de distribución normal para Presion sistólica antes
axs[0, 0].hist(datos_antes, bins=20, density=True, alpha=0.6, color='b')
xmin, xmax = axs[0, 0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_antes, std_antes)
axs[0, 0].plot(x, p, 'k', linewidth=2)
axs[0, 0].set_title('Distribución de Presion sistólica antes')
axs[0, 0].set_xlabel('Presión')
axs[0, 0].set_ylabel('Densidad de probabilidad')

# Histograma y ajuste de distribución normal para Presión sistólica después
axs[0, 1].hist(datos_despues, bins=20, density=True, alpha=0.6, color='g')
xmin, xmax = axs[0, 1].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_despues, std_despues)
axs[0, 1].plot(x, p, 'k', linewidth=2)
axs[0, 1].set_title('Distribución de Presión sistólica después')
axs[0, 1].set_xlabel('Presión')
axs[0, 1].set_ylabel('Densidad de probabilidad')

# Q-Q Plot para Presion sistólica antes
import scipy.stats as stats
stats.probplot(datos_antes, dist="norm", plot=axs[1, 0])
axs[1, 0].set_title('Q-Q Plot de Presion sistólica antes')

# Q-Q Plot para Presión sistólica después
stats.probplot(datos_despues, dist="norm", plot=axs[1, 1])
axs[1, 1].set_title('Q-Q Plot de Presión sistólica después')

plt.tight_layout()
plt.show()
```

Para contestar a la pregunta has de considerar la series de datos 
obtenidas a partir  de las diferencias entre la presión sistólica antes de 
la toma y la presión sistólica al cabo de 60 minutos en el grupo de los 
adultos (contraste de muestras  emparejadas). 

```python
#Ejercicio 3

import pandas as pd
import statsmodels.stats.api as sms
import statsmodels.stats.proportion as proportion
import scipy.stats as stats
from scipy.stats import norm
from scipy.stats import ttest_ind

# Cargar archivo Excel con los datos de Hipertension
df = pd.read_excel("HIPERTENSION.xlsx")

# Filtrar el DataFrame para obtener solo pacientes adultos (grupo 2)
grupo2 = df[df['Grupo'] == 2]

#a)
print("a)")

# Calcular el intervalo de confianza al 95% y al 99%
intervalo_confianza_95 = sms.DescrStatsW(grupo2['Presion sistólica antes']).tconfint_mean()
intervalo_confianza_99 = sms.DescrStatsW(grupo2['Presion sistólica antes']).tconfint_mean(alpha=0.01)

print(f"Intervalo de confianza al 95%: {intervalo_confianza_95}")
print(f"Intervalo de confianza al 99%: {intervalo_confianza_99}")
print("\n")

#b)
print("b)")

grupo1 = df[df['Grupo'] == 1]

# Calcular las medias y las desviaciones estándar de la presión sistólica después de la ingestión del medicamento para cada grupo
media1 = grupo1['Presión sistólica después'].mean()
media2 = grupo2['Presión sistólica después'].mean()
desviacion1 = grupo1['Presión sistólica después'].std()
desviacion2 = grupo2['Presión sistólica después'].std()

# Calcular los tamaños de muestra y los grados de libertad
n1 = len(grupo1)
n2 = len(grupo2)
grados_libertad = n1 + n2 - 2

# Calcular el intervalo de confianza para la diferencia de medias utilizando la distribución normal estándar (Z)
diferencia_medias = media2 - media1
error_estandar = ((desviacion1 ** 2) / n1 + (desviacion2 ** 2) / n2) ** 0.5
z_critical = norm.ppf(0.975)  # 0.975 para un intervalo de confianza del 95%
intervalo_confianza = (diferencia_medias - z_critical * error_estandar, diferencia_medias + z_critical * error_estandar)

# Realizar la prueba t de dos muestras independientes
t_statistic, p_value = ttest_ind(grupo1['Presión sistólica después'], grupo2['Presión sistólica después'], equal_var=False)

# Imprimir los resultados
print("Intervalo de confianza para la diferencia de medias:", intervalo_confianza)
print("Valor p de la prueba t:", p_value)

# Interpretar los resultados
if p_value < 0.05:
    print("Hay evidencia estadística para rechazar la hipótesis nula. La presión sistólica media después de la ingesta del medicamento es diferente entre adultos y jóvenes.")
else:
    print("No hay suficiente evidencia estadística para rechazar la hipótesis nula. No se puede concluir que la presión sistólica media después de la ingesta del medicamento sea diferente entre adultos y jóvenes.")


#c)

print("c)")
# Obtener la presión sistólica antes de la toma del medicamento
presion_antes = df['Presion sistólica antes']

# Calcular la proporción de la población con presión sistólica >= 130 mmHg
proporcion_hipertension = np.mean(presion_antes >= 130)

# Calcular el intervalo de confianza al 99% para la proporción de la población con hipertensión
intervalo_confianza = proportion.proportion_confint(count=np.sum(presion_antes >= 130), nobs=len(presion_antes), alpha=0.01, method='normal')

print(f"Proporción de la población con presión sistólica >= 130 mmHg: {proporcion_hipertension}")
print(f"Intervalo de confianza al 99% para la proporción de hipertensión: {intervalo_confianza}")
print("\n")
# Contraste de hipótesis
proporcion_hipertension_esperada = 0.30
estadistico_z, p_valor = proportion.proportions_ztest(count=np.sum(presion_antes >= 130), nobs=len(presion_antes), value=proporcion_hipertension_esperada)

print(f"Estadístico Z: {estadistico_z}, Valor p: {p_valor}")
print("\n")
nivel_significacion = 0.05
if p_valor < nivel_significacion:
    print("Se rechaza la hipótesis nula. El porcentaje de la población con presión sistólica >= 130 mmHg es distinto de 30%.")
else:
    print("No se rechaza la hipótesis nula. No hay suficiente evidencia para afirmar que el porcentaje de la población con presión sistólica >= 130 mmHg sea distinto de 30%.")
    
print("\n")

#d)

print("d)")
# Filtrar el DataFrame para obtener solo pacientes adultos (grupo 2)
grupo2 = df[df['Grupo'] == 2]

# Obtener las diferencias entre la presión sistólica antes y después del medicamento
diferencias_presion = grupo2['Presión sistólica después'] - grupo2['Presion sistólica antes']

# Realizar el contraste de hipótesis para muestras emparejadas
# H0: No hay diferencia significativa entre la presión sistólica antes y después del medicamento
# H1: Existe diferencia significativa entre la presión sistólica antes y después del medicamento
# Nivel de significación: 0.05
t_stat, p_valor = stats.ttest_rel(grupo2['Presión sistólica después'], grupo2['Presion sistólica antes'])

print(f"Estadístico t: {t_stat}, Valor p: {p_valor}")

# Interpretación
nivel_significacion = 0.05
if p_valor < nivel_significacion:
    print("Se rechaza la hipótesis nula. Existe una diferencia significativa en la presión sistólica antes y después del medicamento en la población del grupo 2.")
else:
    print("No se rechaza la hipótesis nula. No hay suficiente evidencia para afirmar que existe una diferencia significativa en la presión sistólica antes y después del medicamento en la población del grupo 2.")
```
