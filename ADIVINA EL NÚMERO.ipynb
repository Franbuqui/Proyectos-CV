# ADIVINA EL NÚMERO

```python
import random
import openpyxl
import getpass

def generar_numero_aleatorio():
    return random.randint(1, 1000)

def jugar_modo_solitario(intentos):
    numero_a_adivinar = generar_numero_aleatorio()

    for intento in range(1, intentos + 1):
        adivinanza = int(input(f"Introduce tu adivinanza (Intento {intento}): "))

        if numero_a_adivinar == adivinanza:
            print(f"¡Lo adivinaste en el intento {intento}!")
            nombre = input("Escriba su nombre: ")
            guardar_resultado_en_excel(nombre, numero_a_adivinar)
            return True

        elif numero_a_adivinar < adivinanza:
            print("El número buscado es MENOR")

        else:
            print("El número buscado es MAYOR")

    print(f"Lo siento, has agotado tus {intentos} intentos. El número correcto era {numero_a_adivinar}.")
    nombre = input("Escriba su nombre: ")
    guardar_resultado_en_excel(nombre, numero_a_adivinar)
    return False

def jugar_modo_dos_jugadores(intentos, numero_a_adivinar):
    for intento in range(1, intentos + 1):
        adivinanza = int(input(f"Jugador 2 - Introduce tu adivinanza (Intento {intento}): "))

        if numero_a_adivinar == adivinanza:
            print(f"¡Lo adivinaste en el intento {intento}!")
            nombre = input("Escriba su nombre: ")
            guardar_resultado_en_excel(nombre, numero_a_adivinar)
            return True

        elif numero_a_adivinar < adivinanza:
            print("El número buscado es MENOR")

        else:
            print("El número buscado es MAYOR")

    print(f"Lo siento, has agotado tus {intentos} intentos. El número correcto era {numero_a_adivinar}.")
    nombre = input("Escriba su nombre: ")
    guardar_resultado_en_excel(nombre, numero_a_adivinar)
    return False

def guardar_resultado_en_excel(nombre, numero):
    excel_document = openpyxl.load_workbook("C:\\Users\\Fran\\Desktop\\Master\\Temarios python(basico)\\Tarea 1\\Tarea 1.xlsx")
    hoja = excel_document['Numeros']
    hoja.append([nombre, numero])
    excel_document.save("C:\\Users\\Fran\\Desktop\\Master\\Temarios python(basico)\\Tarea 1\\Tarea 1.xlsx")

def mostrar_estadisticas():
    excel_document = openpyxl.load_workbook("C:\\Users\\Fran\\Desktop\\Master\\Temarios python(basico)\\Tarea 1\\Tarea 1.xlsx")
    hoja = excel_document['Numeros']
    Valores = hoja.iter_rows(min_row=1, max_col=2, values_only=True)
    
    for fila in Valores:
        for celda in fila:
            if celda is not None:
                print(celda, end='\t')  
            else:
                break  
        print()  

def main():
    while True:
        print("\n--Menú--")
        print("1. Partida modo solitario \n2. Partida 2 Jugadores \n3. Estadística \n4. Salir")
        opcion = input("Selecciona una opción entre 1 y 4: ")

        if opcion == "1":
            print("1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")
            submenu = input("Selecciona entre 1 y 3: ")

            if submenu in ["1", "2", "3"]:
                intentos = 20 if submenu == "1" else 12 if submenu == "2" else 5
                resultado = jugar_modo_solitario(intentos)

        elif opcion == "2":
            print("1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")
            submenu = input("Selecciona entre 1 y 3: ")

            if submenu in ["1", "2", "3"]:
                intentos = 20 if submenu == "1" else 12 if submenu == "2" else 5
                numero_a_adivinar = int(getpass.getpass("Jugador 1 - Introduce el número a adivinar: "))
                resultado = jugar_modo_dos_jugadores(intentos, numero_a_adivinar)

        elif opcion == "3":
            mostrar_estadisticas()

        elif opcion == "4":
            break

        else:
            print("Opción no válida. Por favor, elige una opción correcta.")

main()
```
